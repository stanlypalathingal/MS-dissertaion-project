```python
#Load the dataset weather
df=pd.read_csv("weather.csv")
df
df.loc[df['play']=='no','play']=0
df.loc[df['play']=='yes','play']=1
df.head() 
P,N=df['play'].value_counts()
print("yes :",P)
print("No :",N)
xtrain=['outlook', 'temperature', 'humidity', 'windy']
ytrain=['play']
df=df[xtrain+ytrain]
```

```python
IG = pd.DataFrame(columns=['column', 'Info_gain'])
Gain = pd.DataFrame(columns=['column', 'gain'])
dependent=''.join(ytrain)
```

```python
# Function to calculate information gain
def informationGain(p,n,b,IG):
    if(p==0):
        return 0
    elif(n==0):
        return 0    
    info_gain= -(p/(p+n))*math.log2(p/(p+n)) - (n/(p+n))*math.log2(n/(p+n))
    IG=IG.append([{'column':b,'Info_gain':round(info_gain,3)}], ignore_index=True)

    print("information gain of ",b, "is ",info_gain)
    return IG
```

```python
IG=informationGain(P,N,'Total',IG)
IG
```

```python
def find_gain(new_df,Gain,info='Total'):
    for col in range(0,len(new_df.columns)-1):
        b=new_df.columns[col]
        print(b)
        cmn=[]

        cmn=new_df.iloc[:,col].unique()
        ig=[]
        p1=[]
        n1=[]


        for col1 in range(0,len(cmn)):
            p=n=0
            print(cmn[col1])
            s=new_df.groupby(b).get_group(cmn[col1])
            print(s)
            p=len(s[s[dependent]==1])
            n=len(s[s[dependent]==0])    
            print(p)
            print(n)
            ig.append(informationGain(p,n,cmn[col1],IG))
            p1.append(p)
            n1.append(n)
            print("information gain of ",cmn[col1], "is ",ig[col1])

        E=0.0
        for i in range(0,len(cmn)):
            E=E+(((p1[i]+n1[i])/(P+N))*ig[i])

        print("Entropy of ", b, "is ",round(E,3))
        gain=round(IG.loc[IG['column']==info,'Info_gain'] -E,3)
        print("Gain of ",b, "is ",gain)
        Gain=Gain.append([{'column':b,'gain':gain}], ignore_index=True)

#         print(Gain)
        print("***********")  
    
    return Gain
```

```python
Gain=find_gain(df,Gain,info='Total')
print(Gain.sort_values(by=['gain'],ascending=False))
```
