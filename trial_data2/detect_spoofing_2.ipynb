```python
def spoof_detect(a, df1):
    b = df1.groupby(["Sensor", "Type", "Units"])["Sensor"].unique().to_frame(name="1").reset_index()
    b = b.drop("1", 1)
#     print(b)
    c = (b.merge(a, how="inner"))
#     print(c)
    fake = pd.concat([c, b], sort=False)
    return fake.drop_duplicates(keep=False)
```
