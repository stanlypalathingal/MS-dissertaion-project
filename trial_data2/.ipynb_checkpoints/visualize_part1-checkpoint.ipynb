```python
df=pd.read_csv("Sample dataset values1.csv")
# df_s=df_s.drop("Key",1)
df=df.dropna()
print("sample before ",df.shape)
```

```python
def plot_cloumns_individual(columnName,type="bar"):
    x=1
    for i in columnName:
        fig.add_subplot(1,3,x)
#         print(df[i].value_counts())
        df[i].value_counts().plot(kind=type)
        
        plt.title(i)
        plt.xticks(rotation=45)
        x=x+1
```

```python
# plt.subplot(2,1,2)
fig = plt.figure(figsize=(15,5))
plot_cloumns_individual(["Sensor","Type","Units"],"bar")
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
plot_cloumns_individual(["Sensor","Type","Units"],"pie")
plt.show()
```
